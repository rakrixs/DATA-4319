##### Below is an actual example of how Gradient Descent works step by step

```python
#Our original fuction

function = lambda x: ((x - 4)**2) + 3
```

```python
x = np.linspace(0, 8, 1000)

#Plot the curve
plt.plot(x, function(x), color = 'm')
plt.show()
```

```python
#Taking the derivitive of the function

def deriv(x):
    x_deriv = 2*(x-4)
    return x_deriv
```

```python
def step_function(x_new, x_prev, precision, l_r):
    x_list, y_list = [x_new], [function(x_new)]
    while abs(x_new - x_prev) > precision:
        x_prev = x_new
        d_x = - deriv(x_prev) #getting the derivitive of the old x
        x_new = x_prev + (l_r * d_x)
        x_list.append(x_new) #apending values to the new x function for visualization
        y_list.append(function(x_new)) #appending to the new y function for visualization

    print ("Local minimum occurs at: "+ str(x_new))
    print ("Number of steps: " + str(len(x_list)))
    
    plt.plot()
    plt.plot(x,function(x), c="b", linewidth = 1)
    plt.scatter(x_list,y_list,c="m")
    plt.plot(x_list,y_list,c="g")
    plt.xlim([-1,10])
    plt.ylim([0,19])
    plt.title("Gradient descent")
    plt.show()
    
    plt.plot()
    plt.plot(x,function(x), c="b", linewidth = 1)
    plt.scatter(x_list,y_list,c="m")
    plt.plot(x_list,y_list,c="g")
    plt.xlim([-1,4.5])
    plt.ylim([1,19])
    plt.title("Zoomed in Gradient descent")
    plt.show()
```

```python
#implementing gradient descent

step_function(0.05, 0, 0.001, 0.46)
```
