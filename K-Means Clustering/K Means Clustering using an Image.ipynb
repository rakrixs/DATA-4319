# K-Means Clustering

The main objective of the K-Means algorithm is to minimize the distance between points within a cluster centroid.

```python
from skimage import io
import numpy as np
import numpy.matlib
import matplotlib.pyplot as plt
import random as random
from sklearn import cluster
import matplotlib.image as mpimg
```

### Reading in the image 

Using matplotlib we read in the image that we will be clustering. Image segmation is used in image processing framework. This is used to classify the image into diffrent groups. K-means is just one of the diffrent methods use to classify images and one of the most popular methods.

```python
image = mpimg.imread('img/img6.jpg')
plt.figure(figsize = (15,8))
plt.title("Original")
plt.imshow(image)
```

### Converting image into a 2D array


```python
rows = image.shape[0]
columns = image.shape[1]
image1 = image/255
X = image1.reshape(rows*columns,3)
X.shape
```

### K-Means clustering using scikit-learn

While there is diffrent methods for implementing the K-Means algorithm the module sklearn.cluster makes the implementation much easier. In the following code we compute k-means clustering.
    - n_clusters: number of clusters
    - max_iter: Maximum number of iterations
    - k-means++: specifies a procedure to initialize the cluster centers before moving forward the algorithm
    - cluster_centers: coordinates of cluster centers
    - labels: the label of each point

```python
kMeans = cluster.KMeans(n_clusters = 3) #Inizializing the number of cluster points
kMeans.fit(X)
```

```python
pred = kMeans.fit_predict(X)

values = kMeans.cluster_centers_
labels = kMeans.labels_

trans = kMeans.transform(X)
```

```python
plt.figure(figsize = (15,8))
plt.imshow(values[labels].reshape(rows, columns, 3))
plt.show()
```

This is our final image with just using 3 cluster points. The image is still not clear and only visible color other than black is green.

Above, the image shows that works with just 3 cluster points but with our image it needs more cluster points to get the distinct colors seperated


Running a couple of more test to start seeing a clearer image

```python
kMeans = cluster.KMeans(n_clusters = 5).fit(X)
values = kMeans.cluster_centers_
labels = kMeans.labels_
plt.figure(figsize = (15,8))
plt.title('n_clusters = 5')
plt.imshow(values[labels].reshape(rows, columns, 3))
plt.show()
```

```python
kMeans = cluster.KMeans(n_clusters = 10).fit(X)
values = kMeans.cluster_centers_
labels = kMeans.labels_
plt.figure(figsize = (15,8))
plt.title('n_clusters = 10')
plt.imshow(values[labels].reshape(rows, columns, 3))
plt.show()
```

```python
kMeans = cluster.KMeans(n_clusters = 16).fit(X)
values = kMeans.cluster_centers_
labels = kMeans.labels_
plt.figure(figsize = (15,8))
plt.title('n_clusters = 16')
plt.imshow(values[labels].reshape(rows, columns, 3))
plt.show()
```

```python
kMeans = cluster.KMeans(n_clusters = 32).fit(X)
values = kMeans.cluster_centers_
labels = kMeans.labels_
plt.figure(figsize = (15,8))
plt.title('n_clusters = 32')
plt.imshow(values[labels].reshape(rows, columns, 3))
plt.show()
```

At 32 cluster points we seem to have a clearer resolution of the picture and we start seeing the background more composed
